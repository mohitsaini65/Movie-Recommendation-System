```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies['overview'] = movies['overview'].fillna('')
movies = movies.dropna()
```

```python
import ast

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L

movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['cast'] = movies['cast'].apply(lambda x: convert(x)[:3])  # top 3 cast only
movies['crew'] = movies['crew'].apply(convert)
```

```python
movies['tags'] = (
    movies['overview']
    + movies['genres'].apply(lambda x: " ".join(x))
    + movies['keywords'].apply(lambda x: " ".join(x))
    + movies['cast'].apply(lambda x: " ".join(x))
)
```

```python
movies['tags'] = movies['tags'].str.lower()
```

```python
final_movies = movies[['movie_id', 'title', 'tags']]
final_movies.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(final_movies['tags'])
```

```python
tfidf_matrix.shape
```

```python
cosine_sim = cosine_similarity(tfidf_matrix)
```

```python
def recommend(movie_title):
    # get index of the movie
    idx = final_movies[final_movies['title'] == movie_title].index[0]

    # similarity scores for this movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # sort (highest similarity first)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # take top 10 similar movies (skip the first one because it's the same movie)
    sim_scores = sim_scores[1:11]

    # return movie titles
    movie_indices = [i[0] for i in sim_scores]
    return final_movies.iloc[movie_indices][['title']]
```

```python
recommend("Avatar")
```

```python
recommend("Batman Begins")
```

```python
recommend('Spectre')
```

```python
import numpy as np
import tensorflow as tf

user_ids = ratings['userId'].values
movie_ids = ratings['movieId'].values
rating_vals = ratings['rating'].values.astype(np.float32)

n_users = ratings['userId'].nunique()
n_movies = ratings['movieId'].nunique()

print("Users:", n_users, "Movies:", n_movies)
```

```python
# User & Movie input layers
user_input = tf.keras.Input(shape=(1,))
movie_input = tf.keras.Input(shape=(1,))

# Embeddings
user_embedding = tf.keras.layers.Embedding(n_users + 1, 50)(user_input)
movie_embedding = tf.keras.layers.Embedding(n_movies + 1, 50)(movie_input)

# Flatten
user_vec = tf.keras.layers.Flatten()(user_embedding)
movie_vec = tf.keras.layers.Flatten()(movie_embedding)

# Concatenate
concat = tf.keras.layers.Concatenate()([user_vec, movie_vec])

# Dense Layers
dense1 = tf.keras.layers.Dense(128, activation='relu')(concat)
dense2 = tf.keras.layers.Dense(64, activation='relu')(dense1)

# Output layer
output = tf.keras.layers.Dense(1)(dense2)

# Model
ncf_model = tf.keras.Model([user_input, movie_input], output)
ncf_model.compile(optimizer='adam', loss='mse')

ncf_model.summary()
```

```python
history = ncf_model.fit(
    [user_ids, movie_ids],
    rating_vals,
    epochs=5,
    batch_size=256,
    verbose=1
)
```

```python
preds = ncf_model.predict([user_ids, movie_ids]).reshape(-1)
rmse_ncf = np.sqrt(np.mean((preds - rating_vals)**2))
rmse_ncf
```
